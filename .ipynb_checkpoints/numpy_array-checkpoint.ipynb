```python
# numpy - ndarray

import numpy as np
a = np.array([1,2,3,4])
print(a)   # 리스트 내 요소 구분하는 쉼표가 없음

print(a[1], a[-1])
print(a[1:])


# 한 가지 타입의 데이터만 저장 가능
b = np.array([1,2,'3',4])
print(b)




# 배열
c = np.zeros(10)
d = np.ones(10)
e = np.eye(3)
print(c)
print(d)
print(e)

print(np.arange(3))
print(np.arange(3,7))
print(np.arange(3,7,2))

a = np.arange(1,2, 0.1)   # 1이상 2미만 구간에서 0.1 간격으로 실수 생성
b = np.linspace(1,2,11)   # 1부터 2까지 11개 구간으로 나눈 실수 생성
print(a)
print(b)

# arange는 간격을 지정. linspace는 구간의 개수를 지정

```

```python
import numpy as np

# 초기값이 5인 배열 100개
a = np.zeros(10)+5
print(a)

a = np.linspace(1, 2, 11)
print(np.sqrt(a))   # 배열의 모든 값이 한꺼번에 계산

import matplotlib.pyplot as plt
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))
plt.show()
```

```python
# mask - 어떤 조건에 부합하는 데이터만 선별적으로 저장

import numpy as np
a = np.arange(-5, 5)
print(a)

print(a<0)

print(a[a<0])

mask1 = abs(a) > 3
print(a[mask1])

mask2 = abs(a) % 2 == 0
print(a[mask1+mask2])   # 둘 중 하나의 조건이라도 참
print(a[mask1*mask2])   # 두 가지 조건 모두 참
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(-100, 100, 1000)   # 1000개의 랜덤 값
y = np.random.randint(-100, 100, 1000)   # 1000개의 랜덤 값
size =  100
mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2]
y = y[mask1+mask2]

plt.scatter(x,y,s=size,c=x,cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
