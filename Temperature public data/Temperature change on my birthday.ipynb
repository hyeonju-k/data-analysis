```python
# 최고 기온 데이터 추출

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)    # 헤더 제외

result = []   # 최고 기온 데이터 저장할 리스트

for row in data :
    if row[-1] != '' :
        result.append(float(row[-1]))
        
print(len(result))
plt.figure(figsize=(13,2))   # 그래프의 크기 조절. 단위 inch
plt.plot(result, 'r')   # result 리스트에 저장된 값을 빨간색 그래프로 그리기

plt.show()
```

```python
# 내 생일 6월19일 데이터 추출


import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)    # 헤더 제외

low = []   # 최저 기온 데이터 저장할 리스트
high = []   # 최고 기온 데이터 저장할 리스트

for row in data :
    if row[-1] != '' and row[-2] != '' :
        if 1983 <= int(row[0].split('-')[0]) :   # 1983년 이후 데이터
            if row[0].split('-')[1] == '06' and row[0].split('-')[2] == '19' :
                low.append(float(row[-2]))
                high.append(float(row[-1]))
        
        
plt.figure(figsize=(13,2))   # 그래프의 크기 조절. 단위 inch
plt.plot(high, 'red', label='high')   # high 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.plot(low, 'blue', label='low')   # low 리스트에 저장된 값을 파란색 그래프로 그리기


plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False   # 마이너스 기호 깨짐 방지
plt.title('Temperature change on my birthday')
plt.legend()
plt.show()
```
