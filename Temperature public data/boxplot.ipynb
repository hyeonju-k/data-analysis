```python
# 상자 그래프 - 자료에서 얻어낸 최댓값, 최솟값, 상위 1/4, 2/4(중앙), 3/4에 위치한 값을 보여주는 그래프

# 1월, 8월 최고 기온

import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

jan = []
aug = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08' :
            aug.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))

# plt.boxplot(jan)
# plt.boxplot(aug)

plt.boxplot([aug, jan])
plt.show()
```

```python
# 월별 최고 기온 데이터

import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

# 월별 데이터를 저장할 리스트 month 생성
month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    if row[-1] != '' :
        month[int(row[0].split('-')[1])-1].append(float(row[-1]))
        
plt.boxplot(month)
plt.show()
```

```python
# 8월 일별 기온 데이터

import matplotlib.pyplot as plt
import csv

f = open('seoul.csv')
data = csv.reader(f)
next(data)

day = []
for i in range(31) :
    day.append([])   # day 리스트에 31개의 리스트 생성
    
for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '08' :
            day[int(row[0].split('-')[2])-1].append(float(row[-1]))
            
plt.style.use('ggplot')   # 그래프 스타일 지정 - 회색 격자 무늬 배경, 2/4선 색 변경
plt.figure(figsize=(10,5), dpi=300)   # 그래프 크기 수정
plt.boxplot(day, showfliers=False)   # 아웃라이어 값 생략

plt.show()
```
