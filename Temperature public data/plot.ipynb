```python
import matplotlib.pyplot as plt
plt.title('plotting')
plt.plot([1,2,3,4],[12, 43, 25, 15])          # plot() - 직선 또는 꺾은선 형태의 그래프를 그릴 때 사용
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('color')

plt.plot([10,20,30,40], color='black', linestyle='--', label='dashed')
plt.plot([40,30,20,10], 'g:', label='dotted') #'<색상><선모양>'을 동시에 적는 형태

plt.legend(loc=1)     # 범례의 위치 0~10
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')
plt.plot([10,20,30,40], 'r.', label='circle')   # 빨간색 원형 마커 그래프
plt.plot([40,30,20,10], 'g^', label='triangle up')     # 초록색 삼각형 마커 그래프
plt.legend()
plt.show()
```
