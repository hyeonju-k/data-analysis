```python
# 히스토그램 - 자료의 분포 상태를 직사각형 모양의 막대 그래프로 나타내는 것

import random
import matplotlib.pyplot as plt

dice = []

for i in range(100000) :   # 주사위 굴리는 횟수를 늘리면 빈도가 비슷해짐
    dice.append(random.randint(1,6))   # 1~6 하나의 숫자 무작위

plt.hist(dice, bins=6)   # bins 옵션 - 가로축의 구간 개수를 설정하는 속성
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

jan = []   # 1월 최고기온
aug = []   # 8월 최고기온

for row in data :
    month = row[0].split('-')[1]   # -로 구분된 값 중 2번째 값을 month에 저장
    if row[-1] != '' :
        if month == '01' :
            jan.append(float(row[-1]))
        if month == '08' :
            aug.append(float(row[-1]))
        
plt.hist(aug, bins=100, color='r', label='Aug')
plt.hist(jan, bins=100, color='b', label='Jan')
plt.legend()
plt.show()
```
