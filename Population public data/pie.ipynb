```python
# pie() - 원그래프. 특정 데이터의 비율을 보기 쉽게 표현

import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')   # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233]   # 데이터
label = ['A형', 'B형', 'AB형', 'O형']   # 레이블
plt.axis('equal')   # 동그란 원
plt.pie(size, labels=label, autopct='%.1f%%')   # autopct - 비율 표시
plt.legend()   # 범례 표시
plt.show()

# 데이터 리스트에 저장된 값들이 3시 방향 부터 반시계 방향으로 표현
```

```python
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')   # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233]   # 데이터
label = ['A형', 'B형', 'AB형', 'O형']   # 레이블
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']   # 색 정하기
plt.axis('equal')   # 동그란 원

plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode=(0,0,0.15,0))
# autopct - 비율 표시
# explode 돌출 효과 - 0은 돌출되지 않음. 
plt.legend()   # 범례 표시
plt.show()

# 데이터 리스트에 저장된 값들이 3시 방향 부터 반시계 방향으로 표현
```
