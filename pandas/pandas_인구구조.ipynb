```python
# pandas로 인구 구조 분석하기

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')

# 1. 데이터 읽어오기
df = pd.read_csv('age.csv', encoding='cp949', index_col=0)

# 2. 연령별 인구 수를 비율로 변경, 총 인구수와 연령구간인구수 데이터 삭제
df = df.div(df['2020년12월_계_총인구수'], axis=0)
del df['2020년12월_계_총인구수'], df['2020년12월_계_연령구간인구수']

name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)
df2 = df[a]

df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_value().index[:5]].T.plot()

plt.show()
```
